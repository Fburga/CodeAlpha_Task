### Overview

The Iris dataset is a multiclass classification dataset introduced by Ronald Fisher in 1936. It contains information about the measurement of150 iris flowers from three different species, used to predict the species based on their physical characteristics. It constitutes of 6 columns and 150 rows. These columns contain some key attributes of the flower as follows:

1-Id: It is the unique identifier for each flower entry.

2-Sepal length(cm)

3-Sepal width(cm)

4-Petal length(cm)

5-Petal width(cm)

6-Species: It contains the specie of the Iris flower.

```python
# Let's drop the Id column since it is just a row number
df.drop(['Id'],axis=1,inplace=True)
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder

# Create a LabelEncoder instance for the species such that setosa=0, versicolor=1 & virginica=2
label_encoder = LabelEncoder()

# Fit and transform the 'Species' column
df['Species'] = label_encoder.fit_transform(df['Species'])

df['Species']
```

```python
df
```

```python
# To Split features(A) from target(B)-the species column
A = df.drop(columns=['Species']) #Features
A
```

```python
B = df['Species'] #Target
B
```

```python
# Split into training and test sets(80% for training and 20% for testing)
#Train-test split(for each train, there's a corresponding test)
from sklearn.model_selection import train_test_split
A_train, A_test, B_train, B_test = train_test_split(A, B, test_size=0.2, random_state=42)
```

```python
# Train decision tree using the training data(fit method)
from sklearn.tree import DecisionTreeClassifier, plot_tree #create a Decision Tree classifier
dec_tree = DecisionTreeClassifier(random_state=42)
dec_tree.fit(A_train, B_train)
```

```python
# Building model using decision tree.
#The goal is to split the data into pure groups, where all samples in a leaf belong to one class.
plt.figure(figsize=(16, 10))
plot_tree(dec_tree, 
          feature_names=A.columns, 
          class_names=['Setosa', 'Versicolor', 'Virginica'], 
          filled=True, 
          rounded=True)
plt.title("Decision Tree Visualization - Iris Dataset")
plt.show()
```

```python
# Train Logistic Regression Model
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter=200) #max_iter=200 eusures convergence during training
model.fit(A_train, B_train)
```

```python
# Make predictions
B_pred = model.predict(A_test)
```

```python
# Evaluate the Model for Accuracy, Classification Report and Confusion Matrix using test data
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
print("Accuracy:", accuracy_score(B_test, B_pred))
print("\nClassification Report:\n", classification_report(B_test, B_pred))
print("\nConfusion Matrix:\n", confusion_matrix(B_test, B_pred))
```

### Summary 

This dataset is an ideal starting point for learning classification, data preprocessing, feature analysis and model evaluation. It has key characteristics as follows:
1- Well-structured: The dataset is clean, with no missing values.
2- Balanced classes: Each species has 50 samples.
3- Low dimensionality: Only 4 features which makes it easy to visualize.
4- Linearly separable(especially Iris-setosa).

Insights from the dataset are:
1- Petal length and petal width are the most informative features.
2- Iris-setosa is easily separable from the other two classes.
3-Iris_versicolor and Iris_virginica have some overlap, making them slightly harder to classify.

Both Logistic Regression and Decision Tree models achieved perfect accuracy on the test set. 
All classes (0=setosa, 1=versicolor, 2=virginica) achieved perfect precision, recall, and F1-score = 1.0
Each model produced the same matrix. Each class is perfectly predicted- no misclassifications.
